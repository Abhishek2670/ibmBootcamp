```python
import torch
import torchvision
from torchvision.transforms import functional as F
import matplotlib.pyplot as plt
import cv2
```

```python
# load a pre trained model of faster rcnn
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()
```

```python
# load the image

image = cv2.imread('main.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# convert the image to tensor
image_tensor = F.to_tensor(image).unsqueeze(0)
```

```python
# perform object detection
with torch.no_grad():
    perdications = model(image_tensor)[0]
```

```python
# exctarct results
boxes = perdications[0]['boxes']
labels = perdications[0]['labels']
scores = perdications[0]['scores']
```

```python
# draw bounding boxes on the image
for i in range(len(boxes)):
    if scores[i] > 0.8:  # threshold to filter weak detections
        x1,y1,x2,y2 = boxes[i]
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
```

```python
# show the image with detections
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
